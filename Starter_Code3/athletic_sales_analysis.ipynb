```python
# Import Libraries and Dependencies
import pandas as pd
from pathlib import Path
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
athletic_sales_2020_path = Path('Resources/athletic_Sales_2020.csv')
athletic_sales_2021_path = Path('Resources/athletic_Sales_2021.csv')

Athletic_Sales_2020_df = pd.read_csv(athletic_sales_2020_path)
Athletic_Sales_2021_df = pd.read_csv(athletic_sales_2021_path)
```

```python
# Display the 2020 sales DataFrame
Athletic_Sales_2020_df
```

```python
# Display the 2021 sales DataFrame
Athletic_Sales_2021_df
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
Athletic_Sales_2020_df.info()
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
merged_athletic_data = pd.concat([Athletic_Sales_2020_df, Athletic_Sales_2021_df], axis=0) 
merged_athletic_data.reset_index()
merged_athletic_data
```

```python
# Check if any values are null.
has_null_values = merged_athletic_data.isna().any()
has_null_values
```

```python
# Check the data type of each column
merged_athletic_data.info()
```

```python
# Convert the "invoice_date" to a datetime datatype
datetime_athletic_data = merged_athletic_data['invoice_date'] = pd.to_datetime(merged_athletic_data['invoice_date'])
datetime_athletic_data
```

```python
# Confirm that the "invoice_date" data type has been changed.
datetime_athletic_data.info()
```

### 2. Determine which Region Sold the Most Products

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
grouped_df = merged_athletic_data.groupby(['region', 'state', 'city'])['units_sold'].sum().reset_index()
grouped_df.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)
# Show the top 5 results.
grouped_df.head(5)
```

```python
# Show the number products sold for region, state, and city.
pivot_table = merged_athletic_data.pivot_table(
    values='units_sold',
    index='region',
    columns=['state', 'city'],
    aggfunc='sum'
)

# Rename the "units_sold" column to "Total_Products_Sold"
pivot_table = merged_athletic_data.pivot_table(
    values='units_sold',
    index='region',
    columns=['state', 'city'],
    aggfunc='sum',
)
pivot_table.rename(columns={'units_sold':'Total_Product_Sales'}, inplace=True)
# Show the top 5 results.
pivot_table.head(5)
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
df_filtered = merged_athletic_data[merged_athletic_data['product'] == 'Women\'s Athletic Footwear']
df_filtered
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"

grouped_df = df_filtered.groupby(['retailer', 'region', 'state', 'city'])['units_sold'].sum().reset_index()
grouped_df.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'}, inplace=True)
# Show the top 5 results.
grouped_df.head()
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
pivot_table = merged_athletic_data.pivot_table(
    values='total_sales',
    index='invoice_date'
)

# Optional: Rename the "total_sales" column to "Total Sales"
pivot_table.rename(columns={'total_sales':'Total_Sales'}, inplace=True)

# Show the table.
pivot_table
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
daily_sales = pivot_table.resample('D')['Total_Sales'].sum()

# Sort the resampled pivot table in descending order on "Total Sales".
daily_sales_sorted = daily_sales.sort_values(ascending=False)
daily_sales
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
weekly_sales = pivot_table.resample('W')['Total_Sales'].sum()

# Sort the resampled pivot table in descending order on "Total Sales".
weekly_sales_sorted = weekly_sales.sort_values(ascending=False)
weekly_sales_sorted
```
